```python
from bs4 import BeautifulSoup as bs
import requests
import csv
import pandas as pd
import numpy as np
import re
```

```python
def clean_num(s):
 
    if s is None:
        return s
    return "".join(re.findall('\d*\.?\d+',s))

def get_data(block, tag, class_text):
    try:
        return block.find(tag, class_ = class_text).text
    except AttributeError:
        return ""
    return
```

```python

URL = requests.get(f"https://www.amazon.sa/-/en/s?i=fashion&bbn=16667061031&rh=n%3A16667061031%2Cp_89%3AGUESS%7CSwarovski&dc&language=en&qid=1638297175&rnid=17161923031&ref=sr_pg_1")
src = URL.content
soup = bs(src,"lxml")
#page_limit = 21
#if(page_num > page_limit):
#print("page ended")
#break

#results = []
#for page in results:

jewelry_list = []
for i in (soup.find_all("div", class_ = "sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20")):
    brand = str.strip(get_data(i, "h5", "s-line-clamp-1"))
    j_type = get_data(i, "h2", "a-size-mini a-spacing-none a-color-base s-line-clamp-2")
    #j_type = get_data(i, True, ["a-size-small a-link-child", "a-size-small a-color-base"])
    rating_no_clean = get_data(i, "span", "a-icon-alt")
    rating = clean_num(get_data(i, "span", "a-icon-alt").split(" ")[0]) 
    review_no_clean = get_data(i, "span","a-size-base")
    review_num = clean_num(get_data(i, "span","a-size-base"))
    price_no_clean = get_data(i, "span","a-offscreen")
    price = clean_num(get_data(i, "span", "a-offscreen"))

    headers = ['brand', 'j_type', 'rating_no_clean', 'rating', 'review_no_clean',
          'review_num','price_no_clean','price']
    jewelry_dict = dict(zip(headers,[ brand,
                                j_type,
                                rating_no_clean,
                                rating,
                                review_no_clean, 
                                review_num,
                                price_no_clean,
                                price]))
    jewelry_list.append(jewelry_dict)
    #results.append(jewelry_list)

#page_num +=1
#time.sleep(40)
#print("page swetched")
```
